```python
import numpy as np
from matplotlib import pyplot as plt
from numpy.fft import fft, ifft
import scipy as sp
from scipy.optimize import curve_fit
import lmfit
from lmfit import Model
import pandas as pd
```

```python
filename_prefix = "~/Downloads/data_conv/"
time_file =filename_prefix+"ue_20240929_1061_merge.csv"
df = pd.read_csv(time_file)
```

```python
timelist = df[df.columns[0]].to_numpy()
time = timelist.tolist()
```

```python
onlist = df[df.columns[1]].to_numpy()
on = onlist.tolist()
```

```python
offlist = df[df.columns[2]].to_numpy()
off = offlist.tolist()
```

```python
result = []
end_index = len(on)

for i in range(end_index):
    result.append(on[i]/off[i])
```

```python
plt.plot(time[13:],result[13:])
```

```python

X = fft(result[10:])
fdata = np.fft.rfftfreq(len(X), d=0.1)
yhatdata = abs(np.fft.rfft(result[10:]))
plt.plot(fdata, yhatdata, 'r')
plt.xlabel('Frequency/THz')
plt.ylabel('Amplitude')
plt.xlim(0,2)
plt.ylim(0,1.2)
plt.axvline(x=0.79)
plt.text(0.79,0.02,'0.79',rotation=0)
plt.axvline(x=1.12)
plt.text(1.12,0.02,'1.12',rotation=0)
plt.show()
```
