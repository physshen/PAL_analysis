```python
import numpy as np
from matplotlib import pyplot as plt
from numpy.fft import fft, ifft
import scipy as sp
from scipy.optimize import curve_fit
import lmfit
from lmfit import Model
from scipy import special
```

```python
num=30
    
t=np.linspace (-5, 15, num)


ycoef1 = 5.26109
ycoef2 = 0
ycoef3 = -0.133734
ycoef4 = -0.337725
ycoef5 = 2.44158
ycoef6 = 1.000



I_real = ( ycoef3 + ycoef4*np.exp(-(t-ycoef2)/ycoef5)  ) * 0.5 * (  1 + special.erf(2*np.sqrt(2)*(t-ycoef2)/ycoef1)  ) + ycoef6

zcoef1 = 3.3
zcoef2 = 0
zcoef3 = -0.00134831
zcoef4 = -0.00514638
zcoef5 = 2.44158
zcoef6 = 1.000

q_real = ( zcoef3 + zcoef4*np.exp(-(t-zcoef2)/zcoef5)  ) * 0.5 * (  1 + special.erf(2*np.sqrt(2)*(t-zcoef2)/zcoef1)  ) + zcoef6
```

```python
plt.plot(t,I_real)
plt.plot(t,q_real)
```

```python
def time_trace_int(t, I20, I30, I40, y_factor, z_factor, G_conv, off1, off2):

    yA =-0.25652
    yt0=0
    ytr=0.3578
    yt1=2.5526
    yC =-0.030184

    y = ( yC + yA*np.exp(-(t-off1)/yt1)  ) * 0.5 * (  1 + special.erf(2*np.sqrt(2)*(t-off1)/G_conv)  )*y_factor + 1

    zA =-21.818
    zt0=0
    ztr=0.3578
    zt1=2140.7
    zC =21.339

    z= ( zC + zA*np.exp(-(t-off2)/zt1)  ) * 0.5 * (  1 + special.erf(2*np.sqrt(2)*(t-off2)/G_conv)  )*z_factor + 1
    
    return (1*y+I20*z+I30+I40)/(1+I20+I30+I40)
```

```python
def time_trace_q(t, I20, I30,I40, y_factor, z_factor, G_conv, off1, off2):
   
    yA =-0.25652
    yt0=0
    ytr=0.3578
    yt1=2.5526
    yC =-0.030184

    y = ( yC + yA*np.exp(-(t-off1)/yt1)  ) * 0.5 * (  1 + special.erf(2*np.sqrt(2)*(t-off1)/G_conv)  )*y_factor + 1

    zA =-21.818
    zt0=0
    ztr=0.3578
    zt1=2140.7
    zC =21.339

    z= ( zC + zA*np.exp(-(t-off2)/zt1)  ) * 0.5 * (  1 + special.erf(2*np.sqrt(2)*(t-off2)/G_conv)  )*z_factor + 1
    
    return ((0.644*1*y+0.712*I20*z+0.576*I30+0.780*I40)/(1*y+I20*z+I30+I40))/((0.644*1+0.712*I20+0.576*I30+0.780*I40)/(1+I20+I30+I40))
```

```python
gmodel_int = Model(time_trace_int)

params = gmodel_int.make_params()

#I20>I30,I40>0
params = gmodel_int.make_params(I30={'value': 0.1, 'min': 0, 'max': 1},
                              I40={'value': 0.1, 'min': 0, 'max': 1},
                              y_factor={'value': 1.5, 'min': 0.01, 'max': 3},
                        z_factor={'value': 1.5 , 'min': 0.01, 'max': 3},
                           G_conv={'value': 3.5, 'min': 0.01, 'max': 6},
                           off1={'value': 0, 'min': -2, 'max': 2}, 
                              off2={'value': 0, 'min': -2, 'max': 2}
                           )
params.add('delta1',value = 0.1, min = 0.01,max=1)
params.add('delta2',value = 0.1, min = 0.01,max=1)
params.add('I20',expr='max(I30+delta1,I40+delta2)')
```

```python

y = I_real
x = t
result = gmodel_int.fit(y, params,t=x,method = 'leastsq')
plt.plot(x, y, 'o')
plt.plot(x, result.best_fit, '-', label='cg')
print(result.fit_report())
```

```python
gmodel_q = Model(time_trace_q)
params = gmodel_q.make_params()

#I20>I30,I40>0
params = gmodel_q.make_params(I30={'value': 0.1, 'min': 0, 'max': 1},
                              I40={'value': 0.1, 'min': 0, 'max': 1},
                              y_factor={'value': 1.5, 'min': 0.01, 'max': 3},
                        z_factor={'value': 1.5 , 'min': 0.01, 'max': 3},
                           G_conv={'value': 3.5, 'min': 0.01, 'max': 6},
                           off1={'value': 0, 'min': -2, 'max': 2}, 
                              off2={'value': 0, 'min': -2, 'max': 2}
                           )
params.add('delta1',value = 0.1, min = 0.01,max=1)
params.add('delta2',value = 0.1, min = 0.01,max=1)
params.add('I20',expr='max(I30+delta1,I40+delta2)')
```

```python
y = q_real
x = t
result = gmodel_q.fit(y, params,t=x,method = 'basinhopping')
plt.plot(x, y, 'o')
plt.plot(x, result.best_fit, '-', label='cg')
print(result.fit_report())
```

```python
x = t
y = I_real/50+1j*q_real
```

```python
def time_trace_complex(t, I20, I30, I40, y_factor, z_factor, G_conv, off1, off2):
    return time_trace_int(t, I20, I30, I40, y_factor, z_factor, G_conv, off1, off2)/50+1j*time_trace_q(t, I20, I30, I40, y_factor, z_factor, G_conv, off1, off2)
```

```python
gmodel_complex = Model(time_trace_complex)
params = gmodel_complex.make_params()

#I20>I30,I40>0
params = gmodel_complex.make_params(I30={'value': 0.1, 'min': 0, 'max': 1},
                              I40={'value': 0.1, 'min': 0.001, 'max': 1},
                              y_factor={'value': 1.5, 'min': 0.01, 'max': 3},
                        z_factor={'value': 1.5 , 'min': 0.01, 'max': 3},
                           G_conv={'value': 3.5, 'min': 0.01, 'max': 6},
                           off1={'value': 0, 'min': -2, 'max': 2}, 
                              off2={'value': 0, 'min': -2, 'max': 2}
                           )
params.add('delta1',value = 0.1, min = 0.01,max=1)
params.add('delta2',value = 0.1, min = 0.01,max=1)
params.add('I20',expr='max(I30+delta1,I40+delta2)')
```

```python
result = gmodel_complex.fit(y, params,t=x,method = 'dual_annealing')
plt.plot(x, np.real(y), 'o')

plt.plot(x, np.real(result.best_fit), '-', label='leastsq')

print(result.fit_report())
```
